```python
# Rynkeby
# Prepare filtered frames in folder Filtered_frames
import shutil
from shutil import copyfile

# Amount of xml_files: 
nxml = len([name for name in os.listdir('project_20_data/video1/framesXML')]) # = 2177

# Amount of frames: 
nframes = len([name for name in os.listdir('Video_frames')]) # = 3001

# List of all xml-files
allxml = os.listdir('project_20_data/video1/framesXML')

# Move the frames that do have a xml-file to another folder "Filtered_frames"
for i in range(0,nframes):
  
  # Turn framenumber into 6-numbered string: 
  framestr = f"{i:06}"
  
  for j in range(0,nxml):
    xml = allxml[j]

    # Remove the prefix and suffix from the xml_filename
    prefix = "frame_"
    suffix = ".xml"
    xml = xml[len(prefix):]
    xml = xml[:-len(suffix)]
    
    # If there is a xml-file to the corresponding frame, move to other folder
    if xml == framestr:
      print("there is a match for frame {}!".format(i))

      shutil.copy("{}/Video_frames/frame{}.jpg".format(os.getcwd(),i),"{}/F".format(os.getcwd(),framestr))
      break
```
